```python
import cv2
import serial

arduino = serial.Serial('COM8', 9600)  # Change COM port as needed

def on_trackbar_change1(value):
    arduino.write(f"1 {value}\n".encode())

def on_trackbar_change2(value):
    arduino.write(f"2 {value}\n".encode())

def on_trackbar_change3(value):
    arduino.write(f"3 {value}\n".encode())

def on_trackbar_change4(value):
    arduino.write(f"4 {value}\n".encode())

cv2.namedWindow("Trackbar Example")
cv2.createTrackbar("Servo 1", "Trackbar Example", 0, 180, on_trackbar_change1)
cv2.createTrackbar("Servo 2", "Trackbar Example", 0, 180, on_trackbar_change2)
cv2.createTrackbar("Servo 3", "Trackbar Example", 0, 180, on_trackbar_change3)
cv2.createTrackbar("Servo 4", "Trackbar Example", 0, 180, on_trackbar_change4)

while True:
    if cv2.waitKey(1) == 27:
        break

arduino.close()
cv2.destroyAllWindows()
```

```python
import cv2
import serial

arduino = serial.Serial('COM8', 9600)  # Change COM port as needed

# List to store trackbar values
trackbar_values = [0, 0, 0, 0]

def on_trackbar_change(index, value):
    arduino.write(f"{index + 1} {value}\n".encode())
    trackbar_values[index] = value
    print("Trackbar values:", trackbar_values)

cv2.namedWindow("Trackbar Example")
cv2.createTrackbar("Servo 1", "Trackbar Example", 0, 180, lambda value: on_trackbar_change(0, value))
cv2.createTrackbar("Servo 2", "Trackbar Example", 0, 180, lambda value: on_trackbar_change(1, value))
cv2.createTrackbar("Servo 3", "Trackbar Example", 0, 180, lambda value: on_trackbar_change(2, value))
cv2.createTrackbar("Servo 4", "Trackbar Example", 0, 180, lambda value: on_trackbar_change(3, value))

while True:
    if cv2.waitKey(1) == 27:
        break

arduino.close()
cv2.destroyAllWindows()
```

```python
import cv2

# Function to be called when trackbar value changes
def on_trackbar_change(index, value):
    trackbar_values[index] = value
    print("Trackbar values:", trackbar_values)

# Create a black image
image = cv2.imread("Rajinikanth_cropped.jpg")  # Load your image here
cv2.imshow("Image", image)

# Initialize trackbar values
num_trackbars = 4
initial_value = 50
max_value = 100
trackbar_values = [initial_value] * num_trackbars

# Create a window for the trackbars
trackbar_window_name = "Trackbar Example"
cv2.namedWindow(trackbar_window_name)

# Create trackbars
for i in range(num_trackbars):
    trackbar_name = f"Value {i+1}"
    cv2.createTrackbar(trackbar_name, trackbar_window_name, initial_value, max_value, lambda value, index=i: on_trackbar_change(index, value))

cv2.waitKey(0)
cv2.destroyAllWindows()
```
